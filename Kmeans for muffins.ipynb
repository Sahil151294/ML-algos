```python
#Here we will be trying to predict the recipee for 
#muffins, scones and cupcakes by using the amount of
#sugar butter and flour required by each item
#by KMEANS Algorithm
```

```python
#immporting the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
#importing the datasset
X = pd.read_csv("recipes_muffins_cupcakes_scones.csv")
#X
```

```python
#slicing out the required dataset from the whole data
X1 = X[["Flour","Sugar","Butter"]]
X1
```

```python
#fitting and training the data

#labels give the cluster numbering for the training set itself
#thus can be validated whether clustering was good or not

kmeans = KMeans(n_clusters=3, random_state=0).fit(X1)

kmeans.labels_
```

```python
#predicting the new points
#So now, if we put arbitrary values we can test 
#in which cluster does the values go to
#if we see its fine, we can deploy the algo for the required problem statement
kmeans.predict([[36, 24,7], [56, 6,19],[19,52,3]])
```

```python
#cluster centres being tried

kmeans.cluster_centers_
```

```python
flour = input('Enter the flour...')
sugar = input('Enter the sugar...')
butter = input('Enter the butter...')
x = kmeans.predict([[flour,sugar,butter]])
if x == 1:
    print("it is a muffin recipee")
elif x == 0:
    print("it is a cupcake recipee")
else:
    print("it is a scones recipee") 
```

```python
import pandas as pd
from sklearn.cluster import KMeans
#from sklearn import metrics
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cardamom.csv')
#print(df)
X = df[['Castor oil','Coconut oil','Silicone oil','Linseed oil']]
Y = df[['Sample']]
```

```python
sns.lmplot('Castor oil','Coconut oil',data=df,hue='Sample',palette = 'Set1',fit_reg = False,scatter_kws={'s':70})
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state = 10)


kmeans = KMeans(n_clusters = 4)
kmeans.fit(x_train)
y_kmeans = kmeans.predict(x_test)
y_kmeans
```

```python
sns.lmplot('Castor oil','Coconut oil',data=df,hue='Sample',palette = 'Set1',fit_reg = False,scatter_kws={'s':70})
plt.scatter(x_test[['Castor oil']],x_test[['Coconut oil']])
```
